```python
amazon_reviews.drop(['categories','colors','dateAdded','dateUpdated','dimension','ean','keys'], axis='columns', inplace=True)
```

```python
amazon_reviews['compound'] = ''
amazon_reviews['negative'] = ''
amazon_reviews['neutral'] = ''
amazon_reviews['positive'] = ''
amazon_reviews.head()
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
print('Start calculating sentiment scores:')

amazon_reviews['compound'] = amazon_reviews['reviews.text'].apply(lambda x: sid.polarity_scores(x)['compound'])
print('Compound Done')
amazon_reviews['negative'] = amazon_reviews['reviews.text'].apply(lambda x: sid.polarity_scores(x)['neg'])
print('Negative Done')
amazon_reviews['neutral'] = amazon_reviews['reviews.text'].apply(lambda x: sid.polarity_scores(x)['neu'])
print('Neutral Done')
amazon_reviews['positive'] = amazon_reviews['reviews.text'].apply(lambda x: sid.polarity_scores(x)['pos'])
print('Positive Done')
print('Stop')

amazon_reviews.head()
```

```python
amazon_reviews.drop('reviews.text',axis='columns',inplace=True)
```

```python
amazon_reviews.corr()
```

```python
x = pd.DataFrame(amazon_reviews, columns = ['compound', 'negative','positive'])
y = pd.DataFrame(amazon_reviews, columns = ['reviews.rating'])
```

```python
y
```

```python
y['reviews.rating'].replace({1: 0, 2: 0}, inplace=True)
```

```python
y['reviews.rating'].replace({3: 1,4: 1, 5: 1}, inplace=True)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
y_train['reviews.rating'].value_counts()
```

```python
y_test['reviews.rating'].value_counts()
```

```python
from sklearn.svm import SVC

clf = SVC(kernel = 'linear', class_weight = 'balanced', probability=True) # Linear Kernel
```

```python
#Train the model using the training sets
clf.fit(x_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(y_test,y_pred)
```
